```python
df_shopping2.drop(columns=["CustomerID"],inplace=True)
```

```python
df_shopping2
```

```python
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0
df_shopping2["Card Member"]=df_shopping2["Card Member"].apply(change_string)
df_shopping2
```

```python
df_shopping2.head(20)
```

```python
df_shopping2["Annual Income"]=df_shopping2["Annual Income"] / 1000
df_shopping2.head()
```

```python
# Saving cleaned data
file_path = "shopping_data_cleaned.csv"
df_shopping2.to_csv(file_path, index=False)
```
